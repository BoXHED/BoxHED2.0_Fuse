```python
def df_checks(df_orig, df_final):
    len1 = len(df_orig)
    len2 = len(df_final)
    assert(len1 == len2) 
    print("lengths match")

    assert([i for i in df_orig.columns if i not in df_final.columns] == [])
    print("old column names match.")
    print(f"new columns: {[i for i in df_final.columns if i not in df_orig.columns]}")

    a = pd.Series(df_orig.sort_values('Icustay')['Icustay'].unique())
    b = pd.Series(df_final.sort_values('Icustay')['Icustay'].unique())
    assert(a.equals(b))
    print("Icustays match")

    c = pd.Series(df_orig.sort_values('subject')['subject'].unique())
    d = pd.Series(df_final.sort_values('subject')['subject'].unique())
    assert(c.equals(d))
    print("subjects match")

    df_final_small = df_final.drop([i for i in df_final.columns if i not in df_orig.columns], axis=1)
    view = df_orig.merge(df_final_small, how='outer', indicator=True)
    bad_rows = view[view['_merge'] != 'both']
    assert(bad_rows.empty)
    print("merge outer test passed")
    
    print("ALL TESTS PASSED")

df_checks(round(test_arash,2), round(test_final,2))
df_checks(round(train_arash,2), round(train_final,2))
```
