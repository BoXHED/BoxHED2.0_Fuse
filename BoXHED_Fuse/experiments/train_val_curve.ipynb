```python
import json

with open('./model/epoch_train_losses.json') as f:
    data = json.load(f)
train_losses = list(data)

with open('./model/epoch_val_losses.json') as f:
    data = json.load(f)
val_losses = list(data)

with open('./model_from_ckpt1/epoch_train_losses.json') as f:
    data = json.load(f)
train_losses.append(data)

with open('./model_from_ckpt1/epoch_val_losses.json') as f:
    data = json.load(f)
val_losses.append(data)

```

```python
import matplotlib.pyplot as plt
import numpy as np

epochs = np.arange(1, 11)

# Plot train and validation curves
plt.plot(epochs, train_losses, label='train')
plt.plot(epochs, val_losses, label='validation')

# Add title and axis labels
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')

# Add legend and show plot
plt.legend()
plt.show()
```
